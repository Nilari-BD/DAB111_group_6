### Drop irrelevant features

- The variable `burns calories (per 30 min)_bc` is very similar to the `burns calories (per 30 min)` feature, thus it is considered redundant for the model and removed.
- The variable `pct_carbs` is just the carbohydrate percentage, which is very similar to the `Carbs` variable, thus it is also considered redundant and has been removed.

- The variable `meal_name` is expected to provide information about the name of the meal (e.g., Breakfast, Lunch, Dinner), but it contains only a single value, "Other". The expected values actually appear in the variable `meal_type`. Therefore, `meal_name` is not necessary, as it does not provide any meaningful or significant information.
- The variable `cal_from_macros` is not described by the author. Since its meaning is unclear and there are already several calorie-related variables, it will be removed.
- The variable `pct_HRR` is not described by the author. It appears to represent a percentage of Heart Rate Recovery (HRR), but since its exact meaning is unclear and there is already a related `Avg_BPM` variable, it will be removed.
- The variable `pct_maxHR` is not described by the author. It appears to represent the maximum percentage of Heart Rate Recovery (HRR), but since its exact meaning is unclear and there is already a related `Max_BPM` variable, it will be removed.
- The variable `cal_balance` is not described by the author. Since its meaning is unclear and there are already several calorie-related variables, it will be removed.

### Checking data consistency

While it was being checked whether the variable descriptions matched the data types and values, it was noticed that the variable `Physical_exercise`, which is supposed to indicate the type or frequency of physical activity, contains negative values. This is inconsistent, because frequency cannot be negative, and such an interpretation does not make sense in this context, participants cannot “undo” exercise.

Given the meaning of `Physical exercise`, negative values are not logical in this context. Since these values represent about 45.95% of the data, and because the source or nature of the error is unknown (for example, whether the negatives indicate a comparison with past frequency or are simply data entry mistakes), the decision was made to drop this column. Keeping it as it is would not provide meaningful information for the analysis.

```python
dataset["Physical exercise"].describe()
```

```python
len(dataset[dataset["Physical exercise"] <= 0])
```

```python
print(f"The negative values in Physical exercise represent {len(dataset[dataset["Physical exercise"] <= 0])/len(dataset)*100:.4}% of the data")
```

In the variable `sugar_g`, which represents sugar content in grams per meal, the same issue is present: there are negative values. A meal cannot have negative sugar content, since sugar is something consumed, not produced, by a meal.

Unlike the issue with `Physical_exercise`, the error in `sugar_g` affects only 0.35% of the data, so removing the affected rows would have minimal impact. However, because this variable is not of interest for the analysis, the entire column will be dropped.

```python
dataset['sugar_g'].describe()
```

```python
len(dataset[dataset["sugar_g"] <= 0])
```

```python
print(f"The negative values in sugar_g represent {len(dataset[dataset["sugar_g"] <= 0])/len(dataset)*100:.4}% of the data")
```

And the problem of negative values it's also present in the variable `cholesterol_mg`, which represents cholesterol content in milligrams per meal. A meal cannot contain negative cholesterol content in grams, as cholesterol is a substance found in food (from animal products) and is not something that can be subtracted.

And similarly as `sugar_g`, this error in `cholesterol_mg` represents a small percente of the data (only 0.55%), so we could simply delete the affected rows. However, because this variable is not of interest for the analysis, the entire column will be dropped.

```python
dataset['cholesterol_mg'].describe()
```

```python
len(dataset[dataset["cholesterol_mg"] <= 0])
```

```python
print(f"The negative values in cholesterol_mg represent {len(dataset[dataset["cholesterol_mg"] <= 0])/len(dataset)*100:.4}% of the data")
```

```python
# Drop inconsistent columns
dataset = dataset.drop(columns=["Physical exercise", "sugar_g", "cholesterol_mg"])
```

### Create derived columns

It was noticed that the dataset’s author mentioned the variable `Age` represents the participant’s age in years. However, the data includes decimal values, which suggests that months are also included. Therefore, if a new variable called `Age_year` is created, it will represent the participant’s age rounded to whole years only.

An `Age_group` column wasn't created because the ages range from a minimum of 18 to a maximum of 59, which generally corresponds to adults (or perhaps young adults and adults). Since the difference is not substantial enough to have a significant analytical impact, grouping by age would not add meaningful insight at this stage.

```python
dataset['Age'].describe()
```

```python
dataset['Age_year'] = np.floor(dataset['Age']).astype("Int64")
```

### Transforming columns

It was noticed that the dataset’s author stated that the variable `Experience_Level` should represent the participant’s fitness experience level (1 = Beginner, 2 = Intermediate, 3 = Advanced). However, the data doesn't contain these integer values (1, 2, 3) or their corresponding category labels (Beginner, Intermediate, Advanced), instead, it displays float numbers.

To address this, the variable will be converted into categorical labels rather than keeping numeric codes. This approach will also make it easier to compare `Experience_Level` with other variables, such as `Difficulty_Level`, which use similar categorical descriptors.

```python
dataset['Experience_Level'].describe()
```

```python
dataset['Experience_Level'] = round(dataset['Experience_Level'])
```

```python
dataset['Experience_Level']
```

```python
# Define a mapping dictionary
category_map = {1: 'Beginner', 2: 'Intermediate', 3: 'Advanced'}

# Replace the numbers with their corresponding category names in the same column
dataset['Experience_Level'] = dataset['Experience_Level'].replace(category_map)
```

```python
dataset['Experience_Level'].unique()
```

### Data type conversions

It was also noticed that the dataset’s author stated that:
- The variable `Daily_meals_frequency` should represent the number of meals consumed per day.
- The variable `Sets` should represent the number of sets completed in the exercise.
- The variable `Reps` should represent the number of repetitions per set.

However, the data doesn't contain integer values; instead, it displays float numbers. Therefore, the variables will be converted into integer numbers.

```python
#Before
dataset['Daily meals frequency'].unique()
```

```python
#Before
dataset['Sets'].unique()
```

```python
#Before
dataset['Reps'].unique()
```

```python
# Convert the float column to an integer type
dataset['Daily meals frequency'] = round(dataset['Daily meals frequency']).astype('Int64')
dataset['Sets'] = round(dataset['Sets']).astype('Int64')
dataset['Reps'] = round(dataset['Reps']).astype('Int64')
```

```python
#After applying conversion
dataset['Daily meals frequency'].unique()
```

```python
#After applying conversion
dataset['Sets'].unique()
```

```python
#After applying conversion
dataset['Reps'].unique()
```

Some object-type columns will be converted into categorical variables, since categorical data is generally much more efficient for operations such as grouping, sorting, or filtering.

However, to decide whether an object-type variable should be converted into a categorical variable, the recommendations from the Pandas documentation will be followed, which suggest converting when:
- The string column contains only a few different values
- The lexical order of a variable is not the same as the logical order (e.g., “one”, “two”, “three”)
- As a signal to other Python libraries that this column should be treated as a categorical variable (e.g. to use suitable statistical methods or plot types)

The following variables are stored as strings:

```python
print(dataset.select_dtypes(exclude = 'number').columns)
print(f"There're {len(dataset.select_dtypes(exclude = 'number').columns)} columns stored as a string")
```

First, the dtype (data type) of the columns that are ordinal categories will be changed, and the correct order will be assigned to them:
- `Experience_Level`
- `Difficulty Level`
- `Burns_Calories_Bin`

```python
#Initial order
print(f"Experience Level values: {dataset["Experience_Level"].unique()}")
print(f"Difficulty Level values: {dataset["Difficulty Level"].unique()}")
print(f"Burns Calories Bin values: {dataset["Burns_Calories_Bin"].unique()}")
```

```python
cat_level = CategoricalDtype(categories=["Beginner", "Intermediate", "Advanced"], ordered=True)
dataset["Experience_Level"] = dataset["Experience_Level"].astype(cat_level)

dataset["Difficulty Level"] = dataset["Difficulty Level"].astype(cat_level)

cat_Bin = CategoricalDtype(categories=["Low", "Medium", "High", "Very High"], ordered=True)
dataset["Burns_Calories_Bin"] = dataset["Burns_Calories_Bin"].astype(cat_Bin)
```

```python
#Final order
print(f"Experience Level values: {dataset["Experience_Level"].unique()}\n")
print(f"Difficulty Level values: {dataset["Difficulty Level"].unique()}\n")
print(f"Burns_Calories_Bin values: {dataset["Burns_Calories_Bin"].unique()}")
```

Now, for the remaining 12 variables that are nominal categorical variables, the number of distinct values they contain will be analyzed.

```python
for nunique_values in list(dataset.select_dtypes(include = 'object').columns):
    print(f"{nunique_values}: {dataset[nunique_values].nunique()}")
```

Based on the number of distinct values, there're 6 variables with more than 10 distinctive values (`Name of Exercise`, `Benefit`, `Target Muscle Group`, `Equipment Needed`, `Type of Muscle`, and `Workout`). Therefore, these will remain as object data type.

The remaining variables will be converted to categorical data types:
- `Gender`
- `Workout_Type`
- `meal_type`
- `diet_type`
- `cooking_method`
- `Body Part`

```python
dataset["Gender"] = dataset["Gender"].astype('category')
dataset["Workout_Type"] = dataset["Workout_Type"].astype('category')
dataset["meal_type"] = dataset["meal_type"].astype('category')
dataset["diet_type"] = dataset["diet_type"].astype('category')
dataset["cooking_method"] = dataset["cooking_method"].astype('category')
dataset["Body Part"] = dataset["Body Part"].astype('category')
```

```python
dataset.info()
```

Originally, the dataset contained dtypes are: float64(39), object(15)

Finally, After cleaning and preprocessing, the dtypes are: Int64(4), category(9), float64(25), object(6)

### Standardize column names

Some of the columns are not labeled properly, which will be problematic when modelling. All column names will be converted to lower case and any spaces will be replaced with underscores for consistency. Any white spaces before or after the column names will also be removed.

```python
dataset.columns
```

```python
# assign new names to the columns
columns_mapping = {
    'Weight (kg)': 'weight_kg',
    'Height (m)': 'height_m',
    'Session_Duration (hours)': 'session_duration_h',
    'Water_Intake (liters)': 'water_intake_l',
    'Workout_Frequency (days/week)': 'workout_frequency',
    'Daily meals frequency': 'daily_meals_frequency',
    'Name of Exercise': 'name_of_exercise',
    'Burns Calories (per 30 min)': 'burns_calories_30m',
    'Target Muscle Group': 'target_muscle_group',
    'Equipment Needed': 'equipment_needed',
    'Difficulty Level': 'difficulty_level',
    'Body Part': 'body_part', 
    'Type of Muscle': 'type_of_muscle',
    'Calories': 'calorie_intake'
}
dataset = dataset.rename(columns=columns_mapping)
```

```python
# make column names lower case and also remove any white spaces
# before & after the column names using the strip() function
dataset.columns = dataset.columns.str.lower().str.strip()
```
