```python
import sqlite3
import pandas as pd
```

```python
# Read and load data from new_myExpenses.csv
new_dataset = pd.read_csv("../data collection/new_PersonalExpenses.csv")
```

```python
con = sqlite3.connect("PersonalExpenses_DB.db") #Create database
cursor = con.cursor() # Create a cursor
```

```python
# Insert the data into the table
new_dataset.to_sql("PersonalExpenses_table",con, index = False, if_exists='replace')
```

```python
#Checking code to call rows and headers
cursor.execute("SELECT * FROM PersonalExpenses_table")
rows = cursor.fetchall()
column_headers = [description[0] for description in cursor.description]
```

```python
for row in rows:
        print(row)
```

```python
print("Column Headers:", column_headers)
```

```python
# commit the changes to the database
con.commit()
```

```python
# close the connection
con.close()
```
